```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
```

```python
a = np.arange(0, 5, 0.02)
plt.subplot(2, 1, 1)
plt.plot(a, f(a))

plt.subplot(2, 1, 2)
plt.plot(a, np.cos(2 * np.pi * a), 'r--')
```

## plt文本显示函数
![img.jpg](./imgs/Snipaste_2020-03-29_16-26-57.png)

```python
a = np.arange(0, 5, 0.02)
plt.plot(a, np.cos(2 * np.pi * a), 'r--')

plt.xlabel('time')
plt.ylabel('wave')
plt.title(r'$y = cos(2\pi x)$')
plt.text(2, 1.5, r'$\mu=100$', fontsize=15)
plt.axis([0, 5, -2, 2])
plt.grid(True)
```

```python
a = np.arange(0, 5, 0.02)
plt.plot(a, np.cos(2 * np.pi * a), 'r--')

plt.xlabel('time')
plt.ylabel('wave')
plt.title(r'$y = cos(2\pi x)$')

plt.annotate(r'$y = cos(2\pi x)$', xy=(2, 1), xytext=(3, 1.5), 
             arrowprops=dict(facecolor='black', shrink=0.1, width=2))
plt.axis([0, 5, -2, 2])
plt.grid(True)
```

### plot.pie()

```python
labels = ['frogs', 'dogs', 'hogs', 'logs']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%', shadow=False, startangle=90)
```

```python
labels = ['frogs', 'dogs', 'hogs', 'logs']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%', shadow=False, startangle=90)
plt.axis('equal')
```

### plt.hist()

```python
mu, sigma = 100, 20
a = np.random.normal(mu, sigma, size=100)
plt.hist(a, 20, histtype="stepfilled", alpha=0.75)
plt.title('Histogram')
```
